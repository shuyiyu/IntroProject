```python
import pandas as pd
import numpy as np
import json

df = pd.read_csv('Yelp_training_data.csv')
df = df[(df['postal_code'] != 85378) & (df['postal_code'] != 89158)]
col = df['population_by_race']
```

```python
import ast
def flatten_race(df, col):
    lst = []
    for i in range(len(col)):
        tmp = pd.DataFrame(ast.literal_eval(col.iloc[i][1:-1])['values'])
        tmp['percent'] = tmp['y']/tmp['y'].sum()
        lst.append(np.array(tmp['percent']))
    return lst
lst = flatten_race(df, col)
df_race = pd.DataFrame(lst, columns=np.array(pd.DataFrame(ast.literal_eval(col.iloc[0][1:-1])['values'])['x'].values))
```

```python
df = pd.concat([df, df_race], axis=1)
# df_race
df
```
